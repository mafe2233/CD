```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Diretório')
print(os.getcwd())
```

```python
dados = pd.read_table("pes2013A.txt", sep = " " , encoding = "windows-1252" )
n = dados.shape[0]
```

```python
dados.shape[0]
```

```python
dados.REGIAO = dados.UF
dados.UF = dados.UF.astype('category')
dados.UF.cat.categories = ('RO','AC','AM','RR','PA','AP','TO','MA','PI','CE','RN','PB','PE','AL','SE','BA','MG','ES','RJ','SP','PR','SC','RS','MS','MT','GO','DF')
# Parte Pedro
dados.V0302 = dados.V0302.astype('category')
dados.V0302.cat.categories = ('Masculino', 'Feminino')
dados.V0404 = dados.V0404.astype('category')
dados.V0404.cat.categories = ('Indígena', 'Branca', 'Preta', 'Amarela', 'Parda', 'Sem declaração')
# Parte Isabella
dados.V6002 = dados.V6002.astype('category')
dados.V4802 = dados.V4802.astype('category')
# Parte Thiago
dados.V4011 = dados.V4011.astype('category')
dados.V4011.cat.categories = ('Casado(a)', 'Desquitado(a) ou separado(a) judicialmente', 'Divorciado(a)', 'Viúvo(a)', 'Solteiro(a)')
dados.V6003 = dados.V6003.astype('category')
dados.V6003.cat.categories = ('Regular do ensino fundamental', 'Regular do ensino médio', 'Educação de jovens e adultos ou supletivo do ensino fundamental', 'Educação de jovens e adultos ou supletivo do ensino médio', 'Superior de graduação', 'Alfabetização de jovens e adultos', 'Creche', 'Classe de alfabetização - CA', 'Maternal, jardim de infância etc.', 'Pré-vestibular', 'Mestrado ou doutorado')
```

```python
dados.REGIAO[(dados.UF == 'GO') | (dados.UF == 'MT') | (dados.UF == 'MS') | (dados.UF == 'DF')]  = "Centro Oeste"
dados.REGIAO[(dados.UF == 'AC') | (dados.UF == 'AM') | (dados.UF == 'AP') | (dados.UF == 'PA') | (dados.UF == 'RR') | (dados.UF == 'RO') | (dados.UF == 'TO')]  = "Norte"
dados.REGIAO[(dados.UF == 'RS') | (dados.UF == 'PR') | (dados.UF == 'SC')]  = "Sul"
dados.REGIAO[(dados.UF == 'ES') | (dados.UF == 'MG') | (dados.UF == 'RJ') | (dados.UF == 'SP')]  = "Sudeste"
dados.REGIAO[(dados.UF == 'AL') | (dados.UF == 'BA') | (dados.UF == 'CE') | (dados.UF == 'MA') | (dados.UF == 'PI') | (dados.UF == 'PE') | (dados.UF == 'PB') | (dados.UF == 'RN') | (dados.UF == 'SE')]  = "Nordeste"
```

```python
print('Tratamento dos dados da variável V0302 (Gênero)')
```

```python
print('Frequências relativas por sexo:')
ut1 = ((dados.V0302.value_counts()/n*100).round(decimals=5))
print(ut1,'\n')
```

```python
plot = ut1.plot(kind='bar',title='Generos',figsize=(6, 6),color=('b','g'))
dados.V0302.cat.categories = ('Masculino', 'Feminino')
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('')
```

```python
plot = ut1.plot(kind='pie',title='Genero',autopct='%.1f',figsize=(6, 6))
```

```python
print('Tabela Cruzada GÊNERO x REGIÃO:')
ct1=pd.crosstab(dados.REGIAO,dados.V0302,margins=True,rownames=['Região'], colnames=['Gênero'])
ct1f = ct1.fillna(0)
ct1f
```

```python
RedeBarras = pd.crosstab(dados.REGIAO, dados.V0302, normalize=True)
RedeBarras.plot(kind='bar', color=('b', 'g'))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
print('Tratamento dos dados da variável V0404 (Etnia)')
```

```python
print('Frequências relativas por Etnia:')
ut2 = ((dados.V0404.value_counts()/n*100).round(decimals=1))
print(ut2,'\n')
```

```python
plot = ut2.plot(kind='bar',title='Etnia',figsize=(6, 6),color=('b','g','r','black','c','y'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('')
```

```python
plot = ut2.plot(kind='pie',title='Raças',autopct='%.1f',figsize=(6, 6))
```

```python
print('Tabela Cruzada ETNIAS x ESTADO:')
ct2=pd.crosstab(dados.REGIAO,dados.V0404,margins=True,rownames=['Região'], colnames=['Etnias'], normalize=True)
ct2f = ct2.fillna(9)
ct2f
```

```python
RedeBarras = pd.crosstab(dados.REGIAO, dados.V0404, normalize=True)
RedeBarras.plot(kind='bar', color=('b','g','r','black','c','y'))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
print('Traramento de dados da variável 6003 (Curso em que frequenta)')
```

```python
print('Frequências relativas por curso que frequenta:')
ut3 = (dados.V6003.value_counts()/n*100).round(decimals=3)
print(ut3,'\n')
```

```python
plot = ut3.plot(kind='bar',title='Cursos',figsize=(6, 6),color=('b','g','y','r','c','orange','m','w','violet','black'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Cursos')
```

```python
plot = ut3.plot(kind='pie',title='Cursos',autopct='%.1f',figsize=(10, 10))
```

```python
print('Tabela Cruzada CURSO x ESTADO:')
ct3=pd.crosstab(dados.REGIAO,dados.V6003, margins=True,rownames=['Região'], colnames=['Curso'], normalize=True)
ct3
```

```python
RedeBarras = pd.crosstab(dados.REGIAO, dados.V6003)
RedeBarras.plot(kind='bar', color=('b','g','y','r','c','orange','m','w','violet','black'),figsize=(10, 10))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
print('Tratamento dos dados da variável V4011 (Estado civil)')
```

```python
print('Frequências relativas por estado civil')
ut4 = (dados.V4011.value_counts()/n*100).round(decimals=3)
print(ut4,'\n')
```

```python
plot = ut4.plot(kind='bar',title='Estado civil',figsize=(6, 6),color=('b','g','r','y','black'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('')
```

```python
plot = ut4.plot(kind='pie',title='Estado civil',autopct='%.1f',figsize=(8, 8))
```

```python
print('Tabela ESTADO CIVIL x ESTADO:')
ct4=pd.crosstab(dados.REGIAO,dados.V4011, margins=True,rownames=['Região'], colnames=['Estado civil'], normalize=True)
ct4
```

```python
RedeBarras = pd.crosstab(dados.REGIAO, dados.V4011)
RedeBarras.plot(kind='bar', color=('b','g','r','y','black'))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
