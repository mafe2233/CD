```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = 0.6
gamma = stats.gamma.rvs(a, loc =1,scale = 1, size = 10000) 
plt.hist(gamma)
```

```python
mug = np.mean(gamma)
varg = np.var(gamma)

print(mug)
print(varg)
```

```python
n = 2
b = np.random.choice(gamma,n)
b
```

```python
np.mean(b)
```

```python
lista = []
for i in range(10000):
    c = np.random.choice(gamma,n)
    med = np.mean(c)
    lista.append(med)
    
lista
```

```python
plt.hist(lista)
```

```python
mul = np.mean(lista)
varl = np.var(lista)
mul,varl
```
