# Resposta item 4:

O teorema do limite central, TLC, constata que se pequenas amostras de uma população fossem retiradas, tendo essas amostras o mesmo tamanho, a distribuição das médias das referidas amostras são equivalentes a uma distribuição normal. Essa equivalência aumenta com o tamanho das amostras. 

Para provar o referido teorema, escolhemos uma distribuição gamma para servir como população. Feito isso, pegamos, dez mil vezes, dois valores aleatórios de nossa distribuição e calculamos a média entre eles. A distribuição obtida era praticamente igual à original. Todavia após aumentarmos gradualmente o número de valores retirados da população, percebemos que a amostra resultante de suas médias se aproximava de uma distribuição normal.

## Resposta item iv:
As distribuições de X1 e X9 são identicas a X

```python
varx1 = np.var(X1)
varx9 = np.var(X9)

desvpadx1 = varx1**0.5
desvpadx9 = varx9**0.5

CovXY = np.cov(X1,X9)
Corr = CovXY[0][1]/(desvpadx1*desvpadx9)
print(Corr)
```
