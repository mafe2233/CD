# Resposta item 4:

O teorema do limite central, TLC, constata que se pequenas amostras de uma populao fossem retiradas, tendo essas amostras o mesmo tamanho, a distribuio das mdias das referidas amostras so equivalentes a uma distribuio normal. Essa equivalncia aumenta com o tamanho das amostras. 

Para provar o referido teorema, escolhemos uma distribuio gamma para servir como populao. Feito isso, pegamos, dez mil vezes, dois valores aleatrios de nossa distribuio e calculamos a mdia entre eles. A distribuio obtida era praticamente igual  original. Todavia aps aumentarmos gradualmente o nmero de valores retirados da populao, percebemos que a amostra resultante de suas mdias se aproximava de uma distribuio normal.

## Resposta item iv:
As distribuies de X1 e X9 so identicas a X

```python
varx1 = np.var(X1)
varx9 = np.var(X9)

desvpadx1 = varx1**0.5
desvpadx9 = varx9**0.5

CovXY = np.cov(X1,X9)
Corr = CovXY[0][1]/(desvpadx1*desvpadx9)
print(Corr)
```
